# Selecting rows using Boolean selection

```python
#age greater than 60
heart['age'] > 60
```

```python
#age greater than 60
heart[heart['age'] > 60]
```

```python
# age greater than 60 and sex == 1
heart[(heart['age'] > 60) & (heart['sex'] == 1)]
```

isin method

```python
# select sex == 0 and 1 using the isin method
heart[heart['sex'].isin([0, 1])]
```

between method

```python
#age between 55 and 60
heart[heart['age'].between(55, 60)]
```
